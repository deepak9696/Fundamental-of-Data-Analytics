<u>__Discrete distribution__</u>  

<u>__Bernoulli distribution:__</u>

  * Probability of success is fixed, which may or maynot be known.  
  __Example:__ Tossing of coin for one trial  
  
<u>__Binomial distribution:__</u>  
  * Exactly 2 mutually exclusive outcomes of a trial.
  * Used to obtain probability of x success from n trial   
  __Example:__ Tossing of coin for n times

<u>__Continuous distribution__</u>    

<u>__Normal distribution:__</u>  
  * Most commonly used and most important distribution
  * Bell-shaped distribution  
  <img src=https://www.itl.nist.gov/div898/handbook/pmc/section5/gifs/normal.gif width="350">
  
  __Characteristics of Normal distribution__  
  * Location is determined by mean(<span>&#956;</span>) and shape is determined by standard deviation(<span>&#963;</span>).  
  <img src=https://images.slideplayer.com/15/4507977/slides/slide_8.jpg width="400">
  * The variable can take any real number between - <span>&#8734;</span> to + <span>&#8734;</span>.
  * Symmetry around mean  
  <img src=https://www.simplypsychology.org/normal-distribution.jpg?ezimgfmt/rscb19/ng:webp/ngcb19 width="350">

  * __Kurtosis:__ Kurtosis measures flatness of the probability distributions of real valued random variable. It measures the amount of probability in the tails. The value is often compared to the curtosis of the normal distribution, which is equal to 3.  
  * __Central limit theorem :__ Sample mean and sample standard deviation can be used as the representative values for population mean and population standard deviation.  
  * __Law of large number :__ The distribution of any variable can be approximated to a normal distribution as the sample size increases no matter what the shape of the population distribution. Generally, we prefer sample size greater than 30.

## _<span style="color:blue">2. Statistical Inference & Hypothesis Testing</span>_  

<u>__Statistical Inference__</u>

<img src=https://biguru.files.wordpress.com/2015/01/introduction-to-statistical-inference.jpg width="350">

Statistical Inference uses a sample to make inferences about a population, using:  
* Statistical estimation
* Hypothesis testing

<u>__Statistical Estimation__</u> : 
  * Point estimation
  * Interval estimation
  
  __Example :__  
  Lots of biscuits produced in a factory during 10 hours of production :  
  135,121,136,142,138,131,129,136,140,133  
  Sample mean is, x<span>&#772;</span> = 134.1  
  We estimate the population mean to be 134.1 (Point estimation)  
  95% Confidence Interval = 134.1 <span>&#8723;</span> 3.747 (Interval estimation)  
  
<u>__Why Hypothesis testing__</u> ?  
We make an assumption about population and try to validate that assumption by using sample data.  

__Case :__  
  * A market researcher wants to determine if the average shopping amount has changed from the previous year at a supermarket.
  * Mean spent was <span>&#36;</span>260.
  * Random sample of 25 shoppers amount spend is taken.  
  The descriptive statistic is:  
    Average shopping spend/month:  
    Total sample size = 25  
    Mean = <span>&#36;</span>330.6  
    Standard deviation = 154.2
    
  * Reasons for the difference:  
    * Actual difference  
    * By chance
    * Design error
      
<u>__Degrees of freedom(df) :__</u>  
  * Degrees of freedom(df) represents the number of values that are free to vary when computing a statistic(e.g. variance).  
  * The number of degrees of freedom is equal to the total number measurements minus the total number of restrictions on those measurements.  
  __Example :__  
  (i) <u>Without Constraint</u>:  
  There are 5 balloons to give to 5 students. As you start to distribute them, you can only play with 4 different balloons to decide which one to give whom. Once you have exhausted 4 options, you are left with 1 balloon to give the last student.  
  
  Here, degrees of freedom is 4.  
  
  (ii) <u>With Constraint</u>:  
  There is 1 balloon to give to a particular student who has attained highest marks.  
  
  Here, the restriction is 1.  
  In this case, Degrees of freedom(df) = Total number of balloons - Total number of restriction = 5-1-1 = 3  
  
<u>__Hypothesis Testing__</u>  

__Null Hypothesis :__
  * The statement we want to test.
  * Denoted by H<sub>0</sub>.
  
  __Example :__  
   H<sub>0</sub>: Average sale per customer in the supermarket is <span>&#36;</span>260.  
   
__Alternate Hypothesis :__  
  * The hypothesis that challenges null hypothesis. 
  * The reverse of H<sub>0</sub>.  
  * Denoted by H<sub>1</sub>.  
  
  __Example :__  
   H<sub>1</sub>: Average sale per customer in the supermarket is not equal to <span>&#36;</span>260.  
   
* The outcome of a hypothesis test is always 'Reject H<sub>0</sub>' or 'Failed to reject H<sub>0</sub>'.  
* 'Reject H<sub>0</sub>' does not mean 'Accepting H<sub>1</sub>'.  

<u>__Type of errors :__</u>  

| Decision taken               | Fail to Reject H<sub>0</sub> | Reject H<sub>0</sub>  |
|:----------------------------:|:----------------------------:|:---------------------:|
| Reality: H<sub>0</sub> True  | Correct decision             | Type-I error          |
| Reality: H<sub>0</sub> False | Type-II error                | Correct decision      |

  <u>__Type-I error :__</u>  
   It occurs if the null hypothesis is rejected even if it is true.  
   H<sub>0</sub> is incorrectly rejected.  
   
  <u>__Type-II error :__</u>  
   It occurs if the null hypothesis is not rejected even if it is false.  
   H<sub>0</sub> is incorrectly accepted.  
   
   * We try to minimize Type-I error to ensure no true H<sub>0</sub> is rejected.  
   
<u>__Significance level and P-value__</u> : 

P(Type-I error) = Significance level = <span>&#593;</span>  

Confidence level = 100(1-<span>&#593;</span>)% = Confidence that a true null hypothesis will not be rejected.   

P-value = Given the sample data, try to find the likelihood of null hypothesis is true.  

If P-value < <span>&#593;</span> , Reject Null hypothesis  
Otherwise, Failed to Reject Null hypothesis  

__Example :__  
H<sub>0</sub> = Average sale per customer in the Supermarket is <span>&#36;</span>260.  
Sample mean, x<span>&#772;</span> = <span>&#36;</span>330.60   
P-value = 0.12 (Only 12% chance H<sub>0</sub> is true)  (Calulated using Statistic table)

As, P-value > <span>&#593;</span>  

__Conclusion :__ Failed to reject Null hypothesis
