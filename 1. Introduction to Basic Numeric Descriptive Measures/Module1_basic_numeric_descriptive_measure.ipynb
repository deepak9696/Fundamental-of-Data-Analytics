### Population vs. Sample   

__Population :__ Population is the entire collection of data which we wish to analyze,describe or draw conclusions about.    

__Sample :__ Sample is the part of the population that we actually examine in order to gather information and draw valid conclusions about larger group.  

[<img src="https://cdn.scribbr.com/wp-content/uploads/2019/09/population-vs-sample-1.png" width="250"/>](https://cdn.scribbr.com/wp-content/uploads/2019/09/population-vs-sample-1.png)    


__Example:__  
A survey is given to randomly select 30 students from $8^{th}$ grade of a school.      
Here, Sample is Randomly selected 30 students.  
Population is all the students studying in $8^{th}$ grade.  

__Why Sampling__?  

* Cost Effective and Quicker analysis  
__Example:__ Let government want to study the effectiveness of a newly launched program for girls between 15 to 25 years of age.Assume 1 million girls between 15 to 25 years of age.  
$(i)$ How much time required to collect all the data?  
$(ii)$ What will be the cost required to reach out to each individual?  
* Less error and more depth - More data collection means more chance of error.  
* Destructive Tests - Collection of data on entire population may not be practical.

## _$\color{blue}{2.\ Measures\ of\ Location}$_  

__Highlights__  
$(1)$ Mean  
$(2)$ Median  
$(3)$ Mode  
$(4)$ Percentile    

$(1)$ __Mean__ :  
Mean is the most common measure of central tendency.  
The mean of a dataseries with numbers $x_{1}$,$x_{2}$,$x_{3}$,.....,$x_{n}$ is:  
[<img src=https://www.onlinemathlearning.com/image-files/xpopulation-mean.png.pagespeed.ic.I2f0Z8ymSt.webp width="350"/>](https://www.onlinemathlearning.com/image-files/xpopulation-mean.png.pagespeed.ic.I2f0Z8ymSt.webp)  

__Example:__ For data points 14,15,30,40  
Mean = $\frac{(14+15+30+40)}{4}$ = 24.75  

$(2)$ __Median__ :  

* A robust measure of central tendency
* Not affected by extreme values present in the data

Median is the middle number in an ordered array.    
* If the count is odd, median is the middle number.  
__Example:__ 3,5,1,9,8  
Ordered number : 1,3,5,8,9  
Median = 5  
* If the count is even, median is the average of 2 middle number.  
__Example:__ 5,9,14,0,7,3   
Ordered number : 0,3,5,7,9,14  
Median = $\frac{(5+7)}{2}$ = 6  

$(3)$ __Mode__ :  

Mode is the value that occurs most often.
* It is unaffected by extreme data values.
* It can be used for numerical or categorical variables.

$(i)$ There may be several modes.  
__Example:__ 6,6,6,8,9,9,10,12,12,12,13,14  
Here, mode = 6 and mode = 12 (both occur 3 times)

$(ii)$ There may be no modes.  
__Example:__ 1,2,3,4,5,6  
Here, each number occurs only 1 time.  

__Why Three Measure__?  

$(i)$ Mean is affected by extreme values, but median is not.  

__Example:__    
40,50,60,70,$\underline{80}$    Here, mean= 60 and median= 60    
40,50,60,70,$\underline{180}$   Here, mean= 80 and median= 60    
As you see, one extreme data point can drastically change the mean, but not median.  

$(ii)$ Median considers only the relative value, but not actual value.  

__Example:__  
Score of 7 students in 2 term tests shown in the below table:  

|          | Student1   | Student2  | Student3  | Student4  | Student5  | Student6  | Student7  |
|:--------:|:----------:|:---------:|:---------:|:---------:|:---------:|:---------:|:---------:|
| Term 1   | 30         | 35        | 45        | 50        | 55        | 60        | 67        |
| Term 2   | 40         | 44        | 48        | 50        | 60        | 70        | 75        |

As you see, median score for both terms is 50, though 6 out of 7 students scored better in Term 2 than Term 1.  

$(4)$__Percentile__ :  

Percentile provides estimation of proportions of the sorted data that should fall above & below a given value. 

The $p^{th}$ percentile is a value such that  
* At most p % of the observations are less than this value.  
* At most (1 - p) % are greater than this value.  

__Example:__  
* $1^{st}$ percentile: Cuts off lowest 1% of sorted data  
* $98^{th}$ percentile: Cuts off lowest 98% of sorted data  

| Term used     | Number of Splits |
|:-------------:|:----------------:|
| Quartile      | 4                |
| Decile        | 10               |
| Pentile       | 20               |
| Percentile    | 100              |  

## _$\color{blue}{3.\ Introduction\ to\ the\ measures\ of\ variability}$_ 

__Highlight__  
$(1)$ Range  
$(2)$ Inter-Quartile range(IQR) & Semi Inter-Quartile Range(SIQR)  
$(3)$ Variance  
$(4)$ Standard deviation  
$(5)$ Coefficient of variation  

$(1)$ __Range__ :  

* Range is a measure of  variation.
* Difference between the largest and the smallest observations in a data series.

__Example:__  
Consider a data series 10,12,13,14,20  
Smallest observation = 10  
Largest observation = 20  
So, range = 20-10 = 10  

$(2)$ __Inter-Quartile range(IQR) & Semi Inter-Quartile Range(SIQR)__ :  

__Inter-Quartile range(IQR)__ :  
* Known as "Mid-Spread" (Spread in the middle 50%)
* Mathematically, IQR = $3^{rd}$ Quartile - $1^{st}$ Quartile  
   
__Semi Inter-Quartile Range(SIQR)__ :  
* Half of IQR
* Mathematically, SIQR = ($3^{rd}$ Quartile - $1^{st}$ Quartile) / 2

$(3)$ __Variance__ :  
* Variance determines the variation around the mean
* Average squared deviation from mean 

  __2 types of Variance__ :  
  * Sample Variance : Data series coming from sample
  * Population variance : Data series coming from population 
[<img src=https://www.onlinemathlearning.com/image-files/xpopulation-variance.png.pagespeed.ic.GgHE6dx_VA.webp width="350"/>](https://www.onlinemathlearning.com/image-files/xpopulation-variance.png.pagespeed.ic.GgHE6dx_VA.webp)  

$(4)$ __Standard daviation__ :  

Standard deviation measures the spread(dispersion) of a data series.  
* Mathematically, Standard deviation = $\sqrt{(Variance)}$
* Has the same units as original data

$(5)$ __Coefficient of variation__ :  

Coefficient of variation(C.V) is a measure of relative variation.  
* Represented in Percentage(%)
* Represents Variation relative to Mean in the data.
* Used to compare 2 or more datasets measured in different units.
* Sensitive to outliers(extreme values) present in the data.

Mathematically,  
[<img src=https://www.statisticshowto.com/wp-content/uploads/2010/03/cv.jpg width="200"/>](https://www.statisticshowto.com/wp-content/uploads/2010/03/cv.jpg)

__Example:__    
A company has Section A(20 employees) & Section B(30 employees).  
* Average weekly wages are \\$550(Section A) & \\$200(Section B).
* Standard deviations are 7(Section A) & 9(Section B).

__Solution:__  
$(i)$ Total wage bill(Section A) = 20 $\times$ 550 = \\$11000    
$(ii)$ Total wage bill(Section B) = 30 $\times$ 200 = \\$6000  
Section A is larger in wage bill.  
Coefficient of variance for Section A = $\frac{7}{550}$ $\times$ 100 = 1.27%  
Coefficient of variance for Section B = $\frac{9}{200}$ $\times$ 100 = 4.5%  

__Conclusion:__ There is a greater variability in the wages of Section B.  


## _$\color{blue}{4.\ Basic\ Measures\ of\ Bivariate\ Analysis}$_ 

__Highlights__  

$(1)$ Covariance  
$(2)$ Pearson Correlation  
$(3)$ Spearman Rank Correlation  

$(1)$ __Covariance__ :  
Covariance is used to measure how two random variables(X,Y) vary in relation to each other.
  
[<img src=https://cdn1.byjus.com/wp-content/uploads/2019/02/Covariance-Formula.png width="300"/>](https://cdn1.byjus.com/wp-content/uploads/2019/02/Covariance-Formula.png)

* Positive Covariance: Variables move in same direction
* Negative Covariance: Variables move in opposite direction
* Values range from - $\infty$ to + $\infty$.
* Higher covariance implies highly inter-related variables.
* For independent variables, Covariance is 0.

__Example:__  
Compute Covariance of stock returns of Stocks A & B.  

| Day     | Stock A | Stock B |
|:-------:|:-------:|:-------:|
| 1       | \\$20   | \\$30   |
| 2       | \\$27   | \\$42   |
| 3       | \\$21   | \\$49   |
| 4       | \\$14   | \\$41   |

__Solution:__  

Step 1: Compute return from Stock price  
Return from Stock A on day 2 = $\frac{(27 - 20)}{20}$ = 0.35  
Similarly,  

| Day     | Stock A   | Stock B   |
|:-------:|:---------:|:---------:|
| 1       | -         | -         |
| 2       | 0.35      | 0.4       |
| 3       | -0.22     | 0.17      |
| 4       | -0.33     | -0.16     | 

Step 2: Calculate the average return of the stocks  
Stock A: $\frac{(0.35 + (-0.22) + (-0.33))}{3}$ = -0.666  
Stock B: $\frac{(0.4 + 0.17 + (-0.16))}{3}$ = 0.136  

Step 3: Compute the Covariance  
Covariance = 0.06 (By applying Sample Covariance formula)  

__Conclusion:__ The computed variance is a positive number. They move in a positive direction.  

__Drawback of Covariance__ :  
* It measures only directional relationship between two random variable.
* It cannot show the strength of relationship between random variables.

$(2)$ __Pearson Correlation__ :  

Pearson Correlation measures the strength of the linear relationship between two quantitative variables.  
* It is unit free.
* Values range between -1 and +1.

Mathematically,  

[<img src=https://cdn1.byjus.com/wp-content/uploads/2019/06/word-image28.png width="300"/>](https://cdn1.byjus.com/wp-content/uploads/2019/06/word-image28.png)

In terms of Covariance,  

[<img src=https://zerodha.com/varsity/wp-content/uploads/2017/03/Formula-correlation.png width="300"/>](https://zerodha.com/varsity/wp-content/uploads/2017/03/Formula-correlation.png)

* It is also known as Product moment correlation.
* Correlation is different from causation.

__Example:__  

Compute Pearson Correlation of stock returns for stocks A & B.  

| Day     | Stock A   | Stock B   |
|:-------:|:---------:|:---------:|
| 1       | -         | -         |
| 2       | 0.35      | 0.4       |
| 3       | -0.22     | 0.17      |
| 4       | -0.33     | -0.16     | 

__Solution:__  

Step 1: Calculate average return of stocks  
Mean Stock A = -0.066  
Mean Stock B = 0.136  

Step 2: Covariance(Stock A, Stock B) = 0.06 (Already calculated in Covariance example)  

Step 3: Calculate Standard Deviation(S.D) of return of stocks  
S.D(Stock A) = 0.298    
S.D(Stock B) = 0.230  

Step 4: Compute the correlation  
r = $\frac{0.06}{(0.298 \times 0.230)}$ = 0.88  (By applying Correlation coefficient formula)  

__Conclusion:__ As Pearson Correlation value is close to +1, there is a strong positive correlation between the returns of these stocks.  

$(3)$ __Spearman Rank Correlation__ :  

Spearman Correlation is a rank correlation.  
* It is computed using the rank.  
* It measures the degree of monotonic relationship.
* It can be used for ordinal variables.

Mathematically,  
[<img src=https://csweb.bournemouth.ac.uk/statistics/all%20images/Spearman%27sformula%20copy.gif width="300"/>](https://csweb.bournemouth.ac.uk/statistics/all%20images/Spearman%27sformula%20copy.gif)
Where, d = $X_{i}$ - $Y_{i}$ (Difference between 2 variables at $i^{th}$ position)  

* Values range between -1 and +1.  
* Unit free measure

__Example:__  

Scores of five students in Mathematics & Science  

| Student   | Math    | Science  |
|:---------:|:-------:|:--------:|
| 1         | 75      | 82       |
| 2         | 65      | 77       |
| 3         | 83      | 93       |
| 4         | 72      | 85       |
| 5         | 88      | 89       |

__Solution:__  

Step 1: Rank of students  

| Student   | Math    | Science  |
|:---------:|:-------:|:--------:|
| 1         | 3       | 4        |
| 2         | 5       | 5        |
| 3         | 2       | 1        |
| 4         | 4       | 3        |
| 5         | 1       | 2        |

Step 2: Difference between Ranks  

| Student   | Math(X) | Science(Y) | d = $X_{i}$ - $Y_{i}$ |
|:---------:|:-------:|:----------:|:---------------------------------:|
| 1         | 3       | 4          | -1                                |
| 2         | 5       | 5          | 0                                 |
| 3         | 2       | 1          | 1                                 |
| 4         | 4       | 3          | 1                                 |
| 5         | 1       | 2          | -1                                |

Step 3: Calculate the rank correlation using formula  
Here, n = 5  
By calculating, $\sum$ $d^2$ = $(-1)^{2}$ + $(0)^{2}$ + $(1)^{2}$ + $(1)^{2}$ + $(-1)^{2}$ = 4  
Now, $r_s$ = 1 - $\frac{(6 \times 4)}{(5 \times 24)}$ = 0.8 (By applying Rank correlation formula)  

__Conclusion:__ The computed correlation is a high positive number, meaning Mathematics & Science have a good positive monotonic relationship.
