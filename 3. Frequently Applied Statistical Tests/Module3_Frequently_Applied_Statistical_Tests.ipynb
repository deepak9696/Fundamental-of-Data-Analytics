<u>__Equality of Means - One Sample__</u>  

__Null Hypothesis :__  
H<sub>0</sub> : Population mean(<span>&#956;</span>) = Hypothesized mean (<span>&#956;</span><sub>0</sub>)  
  
__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#956;</span> > <span>&#956;</span><sub>0</sub>  (One tailed test)  
H<sub>2</sub>: <span>&#956;</span> <span>&#8800;</span> <span>&#956;</span><sub>0</sub>  (Two tailed test)  

__Assumptions :__  
  * The data is a continuous variable.
  * Data comes from a normal distribution.
  * Observations are independent.  
  
If population variance known : Use one sample Z-test  

<img src=https://miro.medium.com/max/287/1*b7iZyQyP8SJ-W51x_L5ekg.png width="250">

If population variance unknown : Use one sample t-test  

<img src=https://miro.medium.com/max/1400/1*a63Vr-UYiXs0g8PHV9zqEA.jpeg width="400">

__Example :__  

Quality control for a machine manufacturing screws. A random sample of 50 screws is taken.     
Given, Expected average diameter, <span>&#956;</span> = 1.5 cm  and Sample mean, x<span>&#772;</span> = 1.51 cm

Assuming Two-tailed test,  
H<sub>0</sub> : <span>&#956;</span> = 1.5 cm    
H<sub>2</sub> : <span>&#956;</span> <span>&#8800;</span> 1.5 cm  

__Case 1 :__(If population variance known)  

Given, Population Standard deviation(<span>&#963;</span>) = 0.03 cm and level of significance(<span>&#593;</span>) = 0.01    
So, Z = (1.51 - 1.5) / (0.03/<span>&#8730;</span>(50)) = 2.357  
P-value = 0.018 (Calculated using Statistics table)  

As, P-value > <span>&#593;</span>  
__Conclusion :__ Failed reject Null Hypothesis  

__Case 2 :__(If population variance unknown)  

Given, Sample Standard deviation(S) = 0.025 cm and level of significance(<span>&#593;</span>) = 0.01  

So, t = (1.51 - 1.5) / (0.025/<span>&#8730;</span>(50)) = 2.828  
degree of freedom(df) = n-1 = 49  
P-value = 0.007 (Calculated using Statistic table)  

As, P-value < <span>&#593;</span>  
__Conclusion :__ Reject Null Hypothesis  

<u>__Equality of Means - Two Independent Sample__</u>  

__Null Hypothesis :__  
H<sub>0</sub> : <span>&#956;</span><sub>1</sub> = <span>&#956;</span><sub>2</sub>  

__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#956;</span><sub>1</sub> <span>&#8800;</span> <span>&#956;</span><sub>2</sub>  

__<u> 3 Scenarios </u> :__  
  * Population variances known : Use two sample Z-test  
  
  <img src=http://www.stat.yale.edu/Courses/1997-98/101/zstat2.gif width="250">
  * Population variances unknown but equal - Use two sample pooled t-test  
  
  <img src=http://www.stat.yale.edu/Courses/1997-98/101/tstatp.gif width="250">
  <img src=http://www.stat.yale.edu/Courses/1997-98/101/tpool.gif width="250">
  * Population variances unknown and not equal - Use two sample t-test  
  
  <img src=http://www.stat.yale.edu/Courses/1997-98/101/tstat2.gif width="250">

__Example :__  
Performance of employees (Computer Engineers(sample 1) vs. Other Engineers(sample 2))      

__Null Hypothesis :__  
H<sub>0</sub> : <span>&#956;</span><sub>1</sub> = <span>&#956;</span><sub>2</sub>  

__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#956;</span><sub>1</sub> <span>&#8800;</span> <span>&#956;</span><sub>2</sub>  

__Case 1 :__(If population variance known)  
Given, Population variance for sample 1 (<span>&#963;</span><sub>1</sub><sup>2</sup>) = 3  
Population variance for sample 2 (<span>&#963;</span><sub>2</sub><sup>2</sup>) = 2.2  
Sample mean for sample 1 (x<span>&#772;</span><sub>1</sub>) = 6.94  
Sample mean for sample 2 (x<span>&#772;</span><sub>2</sub>) = 5.69  
Sample 1 size (n<sub>1</sub>) = 15    
Sample 2 size (n<sub>2</sub>) = 20  
Level of significance(<span>&#593;</span>) = 0.05

So, Z = (6.94-5.69) / (<span>&#8730;</span>((3/15)+(2.2/20))) = 2.245  
P-value = 0.025 (Calculated using statistic table)  

As, P-value < <span>&#593;</span>  
__Conclusion :__ Reject null hypothesis  

__Case 2 :__(If population variance unknown but equal)  
Given, Sample variance for sample 1 (S<sub>1</sub><sup>2</sup>) = 2.95  
Sample variance for sample 2 (S<sub>2</sub><sup>2</sup>) = 2.16  
Sample mean for sample 1 (x<span>&#772;</span><sub>1</sub>) = 6.94  
Sample mean for sample 2 (x<span>&#772;</span><sub>2</sub>) = 5.69  
Sample 1 size (n<sub>1</sub>) = 15    
Sample 2 size (n<sub>2</sub>) = 20  
Level of significance(<span>&#593;</span>) = 0.05  

Pooled variance(S<sub>p</sub><sup>2</sup>) = ((14 <span>&#215;</span> 2.95)+(19 <span>&#215;</span> 2.16))/33 = 2.656  
Pooled standard deviation(S<sub>p</sub>) = <span>&#8730;</span>(2.656) = 1.63  
t = (6.94-5.69) / (1.63 <span>&#215;</span>(<span>&#8730;</span>((1/15)+(1/20)))) = 2.25    
P-value = 0.015 (Calculated using statistic table)  

As, P-value < <span>&#593;</span>  
__Conclusion :__ Reject null hypothesis  

__Case 3 :__(If population variance unknown and not equal)  
Given, Sample variance for sample 1 (S<sub>1</sub><sup>2</sup>) = 2.95  
Sample variance for sample 2 (S<sub>2</sub><sup>2</sup>) = 2.16  
Sample mean for sample 1 (x<span>&#772;</span><sub>1</sub>) = 6.94  
Sample mean for sample 2 (x<span>&#772;</span><sub>2</sub>) = 5.69  
Sample 1 size (n<sub>1</sub>) = 15    
Sample 2 size (n<sub>2</sub>) = 20  
Level of significance(<span>&#593;</span>) = 0.05  

So, t = (6.94-5.69) / (<span>&#8730;</span>((2.95/15)+(2.16/20))) = 2.264    
P-value = 0.015 (Calculated using statistic table)  

As, P-value < <span>&#593;</span>  
__Conclusion :__ Reject null hypothesis  

<u>__Equality of Means - Two Dependent Samples__</u>  

For two dependent samples, we use Paired t-test.  

__Null Hypothesis :__  
H<sub>0</sub> : <span>&#956;</span><sub>1</sub> = <span>&#956;</span><sub>2</sub>  

__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#956;</span><sub>1</sub> < <span>&#956;</span><sub>2</sub> (one-tailed test)    
H<sub>2</sub>: <span>&#956;</span><sub>1</sub> <span>&#8800;</span> <span>&#956;</span><sub>2</sub> (two-tailed test)  

<img src=https://infostatconsulting.files.wordpress.com/2017/04/img_20170428_175956-1.jpg width="250">

S<sub>d</sub> = <span>&#8730;</span>(<span>&#8721;</span> (d<sub>i</sub> - <span>&#396;</span>)<sup>2</sup>)

<u>__Equality of variance - Two Independent Samples__</u>  

For comparing variance of two independent sample, we use F-test.  

<img src=https://images.slideplayer.com/23/6766509/slides/slide_8.jpg width="350">

__Null Hypothesis :__  
H<sub>0</sub> : <span>&#963;</span><sub>1</sub><sup>2</sup> = <span>&#963;</span><sub>2</sub><sup>2</sup>    

__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#963;</span><sub>1</sub><sup>2</sup> <span>&#8800;</span> <span>&#963;</span><sub>2</sub><sup>2</sup> (two-tailed test)  

__Assumption :__  
  * The data is a continuous variable
  * Data comes from a normal distribution  
  * Observations are independent  
  
__Example :__  

Performance of employees (Computer Engineers(sample 1) vs. Other Engineers(Sample 2))    
Given, Sample variance for sample 1(S<sub>1</sub><sup>2</sup>) = 2.95 (greater)   
Sample variance for sample 2(S<sub>2</sub><sup>2</sup>) = 2.16  
Sample size foe sample 1(n<sub>1</sub>) = 15  
Sample size foe sample 2(n<sub>2</sub>) = 20  
Level of significance(<span>&#593;</span>) = 0.05  

__Null Hypothesis :__  
H<sub>0</sub> : <span>&#963;</span><sub>1</sub><sup>2</sup> = <span>&#963;</span><sub>2</sub><sup>2</sup>    

__Alternate Hypothesis :__  
H<sub>1</sub>: <span>&#963;</span><sub>1</sub><sup>2</sup> <span>&#8800;</span> <span>&#963;</span><sub>2</sub><sup>2</sup> (two-tailed test)  

So, F = 2.95 / 2.16 = 1.37  
degree of freedom(df) = n<sub>1</sub> - 1 (14) and n<sub>2</sub> - 1 (19)  
P-value = 0.26 (Calculated using statistic table)  

As, P-value > <span>&#593;</span>  

__Conclusion :__ Failed to reject null hypothesis  

<u>__Chi-Square Test of Independence__</u>  

<u>__Chi-Square__</u> :  
  * Non-Parametric test of statistical test of statistical significance  
  * Different forms of test available
  * Chi-square test of independence


__Example__ :  

Survey of a recently released movie(Gender vs. Rating) shown in the below table :  

| Rating       | Males | Females | Total |
|:------------:|:-----:|:-------:|:-----:|
| Excellent    | 34    | 32      | 66    |
| Very Good    | 23    | 25      | 48    |
| Good         | 21    | 23      | 44    |
| Not so good  | 17    | 17      | 34    |
| Bad/Very Bad | 10    | 8       | 18    |
| Total        | 105   | 105     | 210   |

So, we want to know that, is their a gender association with rating?    
Level of significance(<span>&#593;</span>) = 0.05  

__Solution :__  

__Null Hypothesis :__  
H<sub>0</sub> : There is no association between gender and rating  
H<sub>1</sub> : There is an association between gender and rating  

__Step 1:__  
Observed frequency(O)  

| Rating       | Males | Females | Total |
|:------------:|:-----:|:-------:|:-----:|
| Excellent    | 34    | 32      | 66    |
| Very Good    | 23    | 25      | 48    |
| Good         | 21    | 23      | 44    |
| Not so good  | 17    | 17      | 34    |
| Bad/Very Bad | 10    | 8       | 18    |
| Total        | 105   | 105     | 210   |

__Step 2:__  
Expected frequency(E) : (Row total <span>&#215;</span> Column total) / Grand total  

| Rating       | Males | Females | Total |
|:------------:|:-----:|:-------:|:-----:|
| Excellent    | (105 <span>&#215;</span> 66)/ 210  = 33    | 33      | 66    |
| Very Good    | 24    | 24      | 48    |
| Good         | 22    | 22      | 44    |
| Not so good  | 17    | 17      | 34    |
| Bad/Very Bad | 9     | 9       | 18    |
| Total        | 105   | 105     | 210   |

__Step 3:__  
Test-Statistic = <span>&#8721;</span> <span>&#8721;</span> (O-E)<sup>2</sup>/E  

where, O = observed frequency  
E = Expected frequency  

degrees of freedom(df) = (rows - 1) <span>&#215;</span> (column - 1)  

So, test-statistic = 0.457 (Calculated using above formula)  
df = (5-1) <span>&#215;</span> (2-1) = 4  
P-value = 0.02  
<span>&#593;</span> = 0.05  

As, P < <span>&#593;</span>  
__Conclusion :__ Reject null hypothesis  

<u>__Parametric test vs. Non-Parametric test__</u>  

|     | Parametric test | Non-Parametric test  |
|:-------:|:------------:|:--------------:|
| Data type  | Interval/Ratio scale data     | Nominal or ordinal data    |
| Distributional assumptions  | Required        | Not required          |
| Sample size  | Specific guidelines to be satisfied        | can work for small data size          |
| Power of discrimination  | High        | Comparatively less          |
| Outliers  | Significantly affected        | Not affected          |
| Example  | Z-test, t-test        | Chi-squared test         |
